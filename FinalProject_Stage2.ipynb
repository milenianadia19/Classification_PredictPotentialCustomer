## Outliers

Menggunakan IQR

Menggunakan Z-score

```python
from scipy import stats
print(f'Jumlah baris sebelum memfilter outlier: {len(df)}')

filtered_entries = np.array([True] * len(df))

for col in ['DurationOfPitch','NumberOfPersonVisiting','NumberOfFollowups','NumberOfTrips','MonthlyIncome']:
    zscore = abs(stats.zscore(df[col])) # hitung absolute z-scorenya
    filtered_entries = (zscore < 3) & filtered_entries # keep yang kurang dari 3 absolute z-scorenya
    
df = df[filtered_entries] # filter, cuma ambil yang z-scorenya dibawah 3

print(f'Jumlah baris setelah memfilter outlier: {len(df)}')
```

Digunakan Z-score karena terlalu banyak data yang harus didrop jika menggunakan kriteria outlier IQR

### Label Encoding

```python
df['TypeofContact_label'] = df['TypeofContact'].astype('category').cat.codes
df['Passport_label'] = df['Passport'].astype('category').cat.codes
df['OwnCar_label'] = df['OwnCar'].astype('category').cat.codes
df['Gender_label'] = df['Gender'].astype('category').cat.codes
df['PreferredPropertyStar_label'] = df['PreferredPropertyStar'].astype('category').cat.codes
df['PitchSatisfactionScore_label'] = df['PitchSatisfactionScore'].astype('category').cat.codes
df['CityTier_label'] = df['CityTier'].astype('category').cat.codes
```

```python
df.head()
```

### One Hot Encoding

```python
col = ['Occupation','ProductPitched','MaritalStatus','Designation']
for cat in col:
    onehots = pd.get_dummies(df[cat], prefix=cat)
    df = df.join(onehots)
    
df.head()
```

## Drop Fitur Menggunakan VIF

https://towardsdatascience.com/how-to-remove-multicollinearity-using-python-4da8d9d8abb2

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df[list(df.columns[2:])]
vif_info = pd.DataFrame()
vif_info['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif_info['Column'] = X.columns
vif_info.sort_values('VIF', ascending=False)
```

Fitur yang akan digunakan untuk modeling adalah fitur dengan VIF < 10 yaitu : MonthlyIncome_std, NumberOfPersonVisiting, NumberOfChildrenVisiting, Age, NumberOfFollowups, CityTier_label, NumberOfTrips_norm, DurationOfPitch_std, PitchSatisfactionScore_label, OwnCar_label, Gender_label, PreferredPropertyStar_label, Passport_label, dan TypeofContact_label.

Berdasarkan artikel di bawah, VIF dengan nilai inf menunjukkan bahwa terdapat korelasi yang sempurna
